Import packages needed

```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

Function with image filter

```python
def median_filter(input_image, filter):
    temp = []

    indexer = filter // 2 # Filter size
    filtered_image = []
    
    #Read input image and convert it 
    img = Image.open("PE09_image.jpg").convert("L")
    img.show()
    
    #Set the image height and image width
    image_height = img.height
    image_width = img.width
      
    filtered_image = np.zeros((image_height,image_width))
    
    #Looping through the image_height
    for i in range(image_height):
        #Looping through the image_width
        for j in range(image_width):

            for z in range(filter):
                if i + z - indexer < 0 or i + z - indexer > image_height - 1:
                    for c in range(filter):
                        temp.append(0)
                else:
                    if j + z - indexer < 0 or j + indexer > image_width - 1:
                        temp.append(0)
                    else:
                        for k in range(filter):
                            temp.append(input_image[i + z - indexer][j + k - indexer])
            #sort 
            temp.sort()
            filtered_image[i][j] = temp[len(temp) // 2]
            temp = []
    return filtered_image
```

```python
#Read input image and convert it 
img = Image.open("PE09_image.jpg").convert("L")
img.show()

#Store the image in an input array
image_arr = np.array(img)

#Removing noise from the image
removed_noise = median_filter(image_arr, 5) 
img = Image.fromarray(removed_noise)

#Display image 
img.show()

plt.imshow(img)
```
